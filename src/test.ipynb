```python
import pandas as pd

df = pd.read_csv("../data/camera.csv")

cols = df.select_dtypes(include="number").columns.to_list()

for col in cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    upper_bound = Q3 + 1.5 * IQR
    lower_bound = Q1 - 1.5 * IQR

    min_v = df[col].min()
    max_v = df[col].max()

    print(f"- colonne = {col}\t- upper_bound = {upper_bound}\t- lower_bound = {lower_bound}\t- min_v = {min_v}\t- max_v = {max_v}\n")
```
